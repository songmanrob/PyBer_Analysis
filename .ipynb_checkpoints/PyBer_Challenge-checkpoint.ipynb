```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df
```

```python
city_drivers_df = pyber_data_df.groupby(["city"])
city_drivers_df.head()
```

```python
# From the merged DataFrame get the total rides, total drivers, and total fares for each city type using the groupby() function on the city type.
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
print(total_rides)
print(total_drivers)
print(total_fares)
```

```python
# Calculate the average fare per ride and the average fare per driver for each city type.
avg_fare_per_ride_urban = total_fares["Urban"]/total_rides["Urban"]
avg_fare_per_driver_urban = total_fares["Urban"]/total_drivers["Urban"]
avg_fare_per_ride_suburban = total_fares["Suburban"]/total_rides["Suburban"]
avg_fare_per_driver_suburban = total_fares["Suburban"]/total_drivers["Suburban"]
avg_fare_per_ride_rural = total_fares["Rural"]/total_rides["Rural"]
avg_fare_per_driver_rural = total_fares["Rural"]/total_drivers["Rural"]
print(avg_fare_per_ride_urban)
print(avg_fare_per_driver_urban)
print(avg_fare_per_ride_suburban)
print(avg_fare_per_driver_suburban)
print(avg_fare_per_ride_rural)
print(avg_fare_per_driver_rural)
```

```python
summary_rides = [total_rides["Rural"],total_rides["Suburban"],total_rides["Urban"]]
summary_rides
```

```python
summary_rides = [total_rides["Rural"],total_rides["Suburban"],total_rides["Urban"]]
summary_drivers = [total_drivers["Rural"],total_drivers["Suburban"],total_drivers["Urban"]]
summary_fares = [total_fares["Rural"],total_fares["Suburban"],total_fares["Urban"]]
summary_avg_fare_per_ride = [avg_fare_per_ride_rural, avg_fare_per_ride_suburban, avg_fare_per_ride_urban]
summary_avg_fare_per_driver = [avg_fare_per_driver_rural, avg_fare_per_driver_suburban, avg_fare_per_driver_urban]
summary_df = pd.DataFrame(summary_rides, index = ["Rural", "Suburban", "Urban"], columns = ["Total Rides"])
summary_df["Total Drivers"] = summary_drivers
summary_df["Total Fares"] = summary_fares
summary_df["Average Fare per Ride"] = summary_avg_fare_per_ride
summary_df["Average Fare per Driver"] = summary_avg_fare_per_driver
summary_df = summary_df.style.format({"Total Rides": "{:,}",
                                      "Total Drivers": "{:,}",
                                      "Total Fares": "${:,.2f}",
                                      "Average Fare per Ride": "${:,.2f}",
                                      "Average Fare per Driver": "${:,.2f}"})
summary_df
```

```python
# Rename columns
pyber_data_df = pyber_data_df.rename(columns={'city':'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'})
pyber_data_df.head()
```

```python
# Create a new DataFrame for fares, and include only the City Type and Fare columns using the copy() method on the merged DataFrame
pyber_data_df2 = pyber_data_df[['Date', 'City Type', 'Fare']].copy()
pyber_data_df2.head()
```

```python
# Set the index to the datetime data type
pyber_data_df2[["Date"]] = pyber_data_df2[["Date"]].apply(pd.to_datetime)
```

```python
# Check to make sure the index is a datetime data type by using the info() method on the DataFrame
pyber_data_df2.info()
```

```python
# Set the index to the Date column
pyber_data_df2 = pyber_data_df2.set_index("Date")
pyber_data_df2.head()
```

```python
# Calculate the sum() of fares by the type of city and date using groupby() to create a Series
# Convert the groupby Series into a DataFrame
# Reset the index
total_fares_by_city_and_date_df = pd.DataFrame(pyber_data_df2.groupby(["City Type", "Date"]).sum()["Fare"].reset_index())
total_fares_by_city_and_date_df
```

```python
# Create a pivot table DataFrame with the Date as the index and columns = 'City Type'. The Fare for each Date should appear in each row.
pivot_df = total_fares_by_city_and_date_df.pivot_table(total_fares_by_city_and_date_df, index="Date", columns="City Type")
pivot_df
```

```python
# Create a new DataFrame from the pivot table DataFrame on the given dates, '2019-01-01':'2019-04-28', using loc
pivot_df2 = pivot_df.loc['2019-01-01':'2019-04-28']
pivot_df2
```

```python
# Create a new DataFrame by setting the DataFrame you created in Step 8 with resample() in weekly bins, and calculate the sum() of the fares for each week in the resampled data
pivot_df2_resampled = pivot_df2.resample('W', label="right").sum()
pivot_df2_resampled
```

```python
# Using the object-oriented interface method, plot the DataFrame you created in Step 9 using the df.plot() function
# Import the style from Matplotlib. Use the FiveThirtyEight (Links to an external site.) graph style.
plt.style.use('fivethirtyeight')
x = np.linspace(0, 10)
# Add a title.
fig, ax = plt.subplots(figsize = (20,5))
ax.set_title("Total Fare by City Type")

# Add x- and y-axis labels according to the final figure.
ax.set_ylabel("Fare ($USD)")

ax.xaxis_date()


# ax.plot(x, pivot_df2_resampled, label='line')
ax.plot(pivot_df2_resampled)
# Save the figure to the “analysis” folder.
# Make the figure size large enough so it’s not too small





# pivot_df2_resampled.plot()
#plt.show()
```
